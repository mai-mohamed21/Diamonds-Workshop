```python
df['cut'] = df['cut'].replace({
    'ideal': 'Ideal',
    'very good': 'Very Good',
    'good': 'Good',
    'fair': 'Fair',
    'premium': 'Premium'
})
df['cut'].unique()
```

```python
df['color'].unique()
```

```python
df = df[df['color'] != '87']
df['color'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_cut = LabelEncoder()
df['cut'] = le_cut.fit_transform(df['cut'])
le_clarity = LabelEncoder()
df['clarity'] = le_clarity.fit_transform(df['clarity'])
df=pd.get_dummies(df, columns=['color'],dtype=int)
```

### Encoding Categorical Features

- **Label Encoding (`LabelEncoder`)**
  - Used for **`ordinal features`** (have a natural order).
  - Applied to:
    - `cut`: has order (Fair < Good < Very Good < Premium < Ideal).
    - `clarity`: has order (I1 < SI2 < SI1 < VS2 < VS1 < VVS2 < VVS1 < IF).

- **One-Hot Encoding (`get_dummies`)**
  - Used for **`nominal features`** (no natural order).
  - Applied to:
    - `color`: no inherent order (D, E, F, G, H, I, J).

### Scaling Methods
| Feature | Distribution | Suggested Scaling |
|---------|--------------|-------------------|
| carat   | Right-skewed | Log or RobustScaler |
| depth   | Normal-ish   | StandardScaler     |
| table   | Right-skewed | Log or RobustScaler |
| price   | Strong right skew | Log or RobustScaler |
| x, y, z | Right-skewed with outliers | RobustScaler |

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
df_scaled = df

# -------- 1. Log transform -------
for col in ['carat', 'price', 'x', 'y', 'z']:
    df_scaled[col] = np.log1p(df_scaled[col]) 

# -------- 2. StandardScaler --------
std_scaler = StandardScaler()
df_scaled[['depth']] = std_scaler.fit_transform(df_scaled[['depth']])

# -------- 3. RobustScaler --------
robust_scaler = RobustScaler()
df_scaled[['table']] = robust_scaler.fit_transform(df_scaled[['table']])
print(df_scaled.head())
```
